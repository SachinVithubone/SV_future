```python
#  ------------------------------------------------------------------------------------------------#
# Iris-data for multiclass clasification problem
#  Iris dataset it used for find the diffrent type of flowers based on their leaf width and length
#  ------------------------------------------------------------------------------------------------#

#  import the required library

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
#  5.Removing all null values row
dataf = dataf.dropna(subset=['petalwidth'])
dataf.info()
```

```python
#   If you have the multipleclass classification the get the dummy varaible as below

#   Add the dummy variable  to Class column

dummy = pd.get_dummies(dataf.Class)


# dataf['Class'].replace(["Iris-setosa","Iris-virginica", "Iris-versicolor"], [0,1,2], inplace=True)

#df.loc[df["Purchased"]== 'No', "Purchased"]= 0
```

```python
#  chekc the dummy variables for Class column


dummy.head(100)
#dummy
```

```python
#  Merge the dummy variable and dataset

merged = pd.concat([dataf, dummy], axis='columns')

merged
```

```python
#  now drop the Class column

final_df = merged.drop(['Class'], axis='columns')
```

```python
#  now all iris dataset into number format
final_df.head()
```

```python
#  ----------------------Modeling for your data-------------------------
#  import the logistic regression lib, its used for logistics regression 

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model
```

```python
#  define the x vaiable

x = final_df.drop(['Iris-versicolor','Iris-virginica', 'Iris-setosa'], axis = 'columns')
x
```

```python
#x
```

```python
#  define the y vaiable
y = dataf.Class
y
```

```python
#  now train the your dataset using fit method

model.fit(x,y)
```

```python
#  now check the predict for any length and width for folwers
#  below four parameters are: (sepallength,sepalwidth,petallength,petalwidth)

model.predict([[4,2,3,1]])
```
